### Análise do PIB versus a qualidade de vida

#### Apresentar o gráfico de Dispersão do logaritmo do PIB (Produto Interno Bruto) per capita pela Expectativa de Vida. Cada ponto do gráfico apresenta o valor do logaritmo do PIB per capita de um país em relação a Expectativa de Vida do país.

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set();
import plotly.express as px
```

```python
url = pd.read_excel('gapminder_final.xlsx')
```

```python
data = pd.DataFrame(url)
data.head()
```

```python
rename_columns = data.rename(
                columns = {
                    "country": "país",
                    "year": "ano",
                    "gdpPercap": "perCap",
                    "lifeExp": "expVida",
                    "pop": "populacao",
                    "continent": "continente" 
                }
)
```

```python
rename_columns.head()
```

```python
df = pd.DataFrame(rename_columns)
df.head()
```

### Algoritmo para plotagem do gráfico

#### Para construir o Scatter plot será utilizada a biblioteca 'plotly.express', é uma biblioteca de visualização de dados do Python, com ela é possível construir gráficos de maneira simples e esteticamente bonitos.

```python
fig = px.scatter(df, x = "perCap", y = "expVida", log_x = True, width = 1000, height = 800)
fig.update_traces(maker = dict(size = 12, line = dict(width = 2)), selector = dict(mode = "makers"))
fig.update_layout(title = "Log do PIB versus Expectativa de vida")
fig.update_xaxes(title = "Log(PIB per capita)")
fig.update_yaxes(title = "Expectativa de vida")
fig.show()
```

#### Podemos observar que quanto maior o PIB, maior a expectativa de vida.

#### O gráfico está com grande poluição visual, por isso iremos adotar os 1000(mil) dados aleatórios.

```python
df_sample = df.sample(n=1000)
df_sample.head()
```

```python
fig = px.scatter(df_sample, x="perCap", y="expVida", hover_name="país", log_x = True, width = 1000, height = 800)
fig.update_traces(marker = dict(size=12, line = dict(width=2)), selector=dict(mode='markers'))
fig.update_layout(title = 'Logaritmo do PIB per capita X Expectativa de vida')
fig.update_xaxes(title = "Log(PIB per capita)")
fig.update_yaxes(title = "Expectativa de vida")
fig.show()
```

#### Ano de 2019

```python
df_2019 = df[df["ano"] == 2019]
fig = px.scatter(df_2019, x = "perCap", y = "expVida", hover_name = "país", log_x = True, size_max = 90, width = 800)
fig.update_traces(marker = dict(size = 12, line = dict(width=2)), selector = dict(mode = 'markers'))
fig.update_layout(title = 'Logaritmo do PIB per capita X Expectativa de vida no ano de 2019')
fig.update_xaxes(title = 'Log (PIB per capita)')
fig.update_yaxes(title = 'Expectativa de vida')
fig.show()
```
